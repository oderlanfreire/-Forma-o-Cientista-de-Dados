# Exemplo de grafico 3D

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
df_orchard = pd.read_csv('../data/orchard.csv', sep=',', encoding='latin')
print(f'linhas: {df_orchard.shape[0]}\ncolunas: {df_orchard.shape[1]}')
df_orchard.head()
```

```python
#figura 3d
figure = plt.figure()
eixo = figure.add_subplot(1,1,1, projection='3d')
eixo.scatter(df_orchard['decrease'],df_orchard['rowpos'],df_orchard['colpos'])
eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')
```
