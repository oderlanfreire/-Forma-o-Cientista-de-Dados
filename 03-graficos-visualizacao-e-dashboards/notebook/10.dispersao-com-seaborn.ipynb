# Grafico de Dispersão usando Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_co2 = pd.read_csv('../data/co2.csv', sep=',', encoding='latin')
print(f'linhas: {df_co2.shape[0]}\ncolunas: {df_co2.shape[1]}')
df_co2.head()
```

```python
#grafico de dispersão utilizando os atributos conc e uptake, agrupando por type
sns.scatterplot(x=df_co2['conc'], y=df_co2['uptake'], hue=df_co2['Type'])
```

```python
#seleção de registros 
mississippi = df_co2.loc[df_co2['Type'] == 'Mississippi']
quebec = df_co2.loc[df_co2['Type'] == 'Quebec']
```

```python
#figura com graficos
plt.figure()
plt.subplot(1, 2, 1)
sns.scatterplot(x=quebec['conc'], y=quebec['uptake'], hue=quebec['Type'])
plt.title('Quebec')
plt.subplot(1, 2, 2)
sns.scatterplot(x=mississippi['conc'], y=mississippi['uptake'], hue=mississippi['Type'])
plt.title('Mississippi')
plt.tight_layout()
```

```python
#seleção de registros refrigerado e não refrigerado
refr = df_co2.loc[df_co2['Treatment'] == 'chilled']
nrefr = df_co2.loc[df_co2['Treatment'] == 'nonchilled']
```

```python
#geração de figura
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(x=refr['conc'], y=refr['uptake'])
plt.title('Chilled')
plt.subplot(1,2,2)
sns.scatterplot(x=nrefr['conc'], y=nrefr['uptake'])
plt.title('Nonchilled')
plt.tight_layout()
```

```python
#carregamento de nova base de dados
df_esoph = pd.read_csv('../data/esoph.csv', sep=',', encoding='latin')
print(f'linhas: {df_esoph.shape[0]}\ncolunas: {df_esoph.shape[1]}')
df_esoph.head()
```

```python
#utilização do catplot para demonstrar a relação entre alcgp e ncontrols
sns.catplot(x= 'alcgp', y='ncontrols', data=df_esoph, jitter= False)
```

```python
#agora a versão com agrupamento pela coluna tobgp
sns.catplot(x= 'alcgp', y='ncontrols', data=df_esoph, col= 'tobgp')
```
