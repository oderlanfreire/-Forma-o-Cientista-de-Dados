# Exemplo de boxplot

```python
#importação de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Carregamento da base de dados
df_trees = pd.read_csv('../data/trees.csv', sep=',', encoding='latin')
print(f'linhas: {df_trees.shape[0]}\ncolunas: {df_trees.shape[1]}')
df_trees.head()
```

```python
#geração de boxplot
#vert para selecionar a posição do boxplot se é vertical ou horizontal
#shofliers é para mostrar pontos de outliers
#notch é para recortar o centro
#patch_artist é para preencher
plt.boxplot(df_trees['Volume'], vert= False, showfliers= False, notch= True, patch_artist=True)
plt.title('Árvore')
plt.xlabel('Volume')
```

```python
#dados por linha
plt.boxplot(df_trees)
plt.title('Árvore')
plt.xlabel('Volume')
```

```python
#boxplot de cada coluna
plt.boxplot(df_trees['Girth'], vert=False)
plt.boxplot(df_trees['Height'], vert=False)
plt.boxplot(df_trees['Volume'], vert=False)
plt.title('Árvores')
plt.xlabel('Dados')
```
