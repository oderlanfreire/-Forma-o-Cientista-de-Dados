# Exemplo de gráfico de densidade

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tree = pd.read_csv('../data/trees.csv', sep=',', encoding='latin')
print(f'linhas: {df_tree.shape[0]}\ncolunas: {df_tree.shape[1]}')
df_tree.head()
```

```python
plt.hist(df_tree.iloc[:, 1], bins=6, color='gray')
```

```python
#histograma com linha de distribuição de frequência, kde é a linha de densidade mas nesse vai ser
#false
sns.histplot(df_tree.iloc[:, 1], kde=False, bins=6, color='gray').set(title="Árvores")
```

```python
#grafico de densidade, kdplot para densidade
sns.kdeplot(df_tree.iloc[:, 1], color='black').set(title='Árvore')
sns.kdeplot(df_tree.iloc[:, 1], fill=True, color='gray')
```

```python
#histograma com densidade(kde)
sns.histplot(df_tree.iloc[:, 1], kde=True, color='gray').lines[0].set_color('Black')
plt.title("Árvores - Histograma com Densidade(KDE)")
```
