# Histograma e densidade no seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_trees = pd.read_csv('../data/trees.csv', sep=',', encoding='latin')
df_trees.head()
```

```python
#histograma com 10 divisões (bins) e com graficos  de densidade (kernel density - suavização)
#stats: count, frequency, density, probability,  percent
sns.histplot(df_trees['Volume'], color='blue', kde=True, stat="density", bins=10).set_title("Volume")
```

```python
df_chicken = pd.read_csv('../data/chicken.csv', sep=',', encoding='latin')
print(f'linhas: {df_chicken.shape[0]}\ncolunas: {df_chicken.shape[1]}')
df_chicken.head()
```

```python
group_feed = df_chicken.groupby(['feed'])['weight'].sum()
group_feed
```

```python
#dataframe de teste de filtro com pandas
df_filtrado = df_chicken.loc[df_chicken['feed'] == 'horsebean']
df_filtrado
```

```python
#grafico de densidade considerando apenas o valor de 'horsebean'
sns.kdeplot(df_chicken.loc[df_chicken['feed'] == 'horsebean'].weight).set(title='Horsebean')
sns.kdeplot(df_chicken.loc[df_chicken['feed'] == 'horsebean'].weight, color='gray', fill=True).lines[0].set_color('black')
```

```python
#histograma considerando apenas o valor casein
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'casein'].weight, kde=True, color='gray').lines[0].set_color('black')
plt.title('casein')
```

```python
#histograma considerando apenas o valor linseed
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'linseed'].weight, kde=True, color='gray').lines[0].set_color('black')
plt.title('linseed')
```

```python
#histograma considerando apenas o valor meatmeal
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'meatmeal'].weight, kde=True, color='gray').lines[0].set_color('black')
plt.title('meatmeal')
```

```python
#histograma considerando apenas o valor soybean
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'soybean'].weight, kde=True, color='gray').lines[0].set_color('black')
plt.title('soybean')
```

```python
#histograma considerando apenas o valor sunflower
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'sunflower'].weight, kde=True, color='gray').lines[0].set_color('black')
plt.title('sunflower')
```

```python
#figura composta pelos graficos
plt.figure()
plt.subplot(3,2,1)
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'horsebean'].weight, kde=True, color='gray').lines[0].set_color('black')
plt.title('horsebean')
plt.subplot(3,2,2)
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'casein'].weight, kde=True, color='gray').lines[0].set_color('black')
plt.title('casein')
plt.subplot(3,2,3)
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'linseed'].weight, kde=True, color='gray').lines[0].set_color('black')
plt.title('linseed')
plt.subplot(3,2,4)
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'meatmeal'].weight, kde=True, color='gray').lines[0].set_color('black')
plt.title('meatmeal')
plt.subplot(3,2,5)
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'soybean'].weight, kde=True, color='gray').lines[0].set_color('black')
plt.title('soybean')
plt.subplot(3,2,6)
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'sunflower'].weight, kde=True, color='gray').lines[0].set_color('black')
plt.title('sunflower')
plt.tight_layout()
#plt.tight_layout é para não haver sobreposição dos graficos
```
