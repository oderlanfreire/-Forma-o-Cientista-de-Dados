# Exemplo de gráfico de dispersão

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tree = pd.read_csv('../data/trees.csv', sep=',', encoding='latin')
print(f'linhas: {df_tree.shape[0]}\ncolunas: {df_tree.shape[1]}')
df_tree.head()
```

```python
#grafico de dispersão considerando o volume e dispersão
plt.scatter(x=df_tree.Girth, y=df_tree.Volume, color='gray', facecolors='none', marker='*')
plt.title('Árvores - Gráfico de dispersão')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
# Grafico de linha considerando o girth e o volume
plt.plot(df_tree.Girth, df_tree.Volume, color='black')
plt.title('Árvore - Gráfico de linha')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
#gráfico de dispersão com jitter (afastamento de dados)
#fit_reg linha de tendencia
sns.regplot(x=df_tree.Girth, y=df_tree.Volume, color='gray', data=df_tree, x_jitter= 0.3, fit_reg=False)
```
