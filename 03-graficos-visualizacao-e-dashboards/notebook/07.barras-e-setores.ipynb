# Exemplo de graficos de barras e setores

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#carregamento da base de dados, contagem de linhas e colunas
df_insect = pd.read_csv('../data/insect.csv', sep=',', encoding='latin')
print(f'linhas: {df_insect.shape[0]}\ncolunas: {df_insect.shape[1]}')
df_insect.head()
```

```python
#agrupamento pela coluna spray, contando e somando os valores
agrupamento = df_insect.groupby(['spray'])['count'].sum()
agrupamento
```

```python
#grafico de barras
agrupamento.plot.bar(color='gray')
plt.xticks(rotation=1)
```

```python
#grafico de barras de colunas (com cores)
agrupamento.plot.bar(color=['green', 'blue', 'red', 'brown', 'orange', 'pink'])
plt.xticks(rotation=1)
```

```python
#agrupamento de pizza
agrupamento.plot.pie()
```

```python
#grafico de pizza com legenda
agrupamento.plot.pie(legend = True)
```
