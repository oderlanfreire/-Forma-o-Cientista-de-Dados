# Exemplo de divisão de tela para mostar gráficos variados

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#carregamento da base de dados
df_trees = pd.read_csv('../data/trees.csv', sep=',', encoding='latin')
print('linhas / colunas')
print(df_trees.shape)
df_trees.head()
```

```python
#dispersão girth com volume
plt.scatter(x=df_trees['Girth'], y=df_trees['Volume'])
```

```python
#dispersão de girth com height
plt.scatter(x=df_trees['Girth'], y=df_trees['Height'])
```

```python
#dispersão de height com volume
plt.scatter(x=df_trees['Height'], y=df_trees['Volume'])
```

```python
#histograma de volume
plt.hist(df_trees['Volume'])
```

```python
#criação de figura para colocar os graficos juntos
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(x=df_trees['Girth'], y=df_trees['Volume'])
plt.subplot(2,2,2)
plt.scatter(x=df_trees['Girth'], y=df_trees['Height'])
plt.subplot(2,2,3)
plt.scatter(x=df_trees['Height'], y=df_trees['Volume'], marker='*')
plt.figure(2,2,4)
plt.hist(df_trees['Volume'])
```
