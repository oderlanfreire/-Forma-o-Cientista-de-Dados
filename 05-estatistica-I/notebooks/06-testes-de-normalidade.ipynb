# Testes de normalidade

```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
#criação da variavel com dados em uma distribuição normal com a função rvs
data = norm.rvs(size = 1000)
data
```

```python
#histograma
plt.hist(data, bins=20)
plt.title('Dados')
```

```python
#geração de grafico para verificar se ele se aproxima de distribuição normal
fig, ax = plt.subplots()
stats.probplot(data, fit=True, plot=ax)
plt.show()
```

```python
#teste de shapiro
stats.shapiro(data)
```

o segundo argumento é o valor de p, não há como rejeitar a hipotese nula, pois o valor está acima do alfa(0.05)

## Execução de testes em dados não normais

```python
data2 = skewnorm.rvs(4, size=1000)
```

```python
#histograma
plt.hist(data2, bins=20)
plt.title('dados')
```

```python
#geração de grafico para verificar se a distribuição é normal

fig, ax = plt.subplots()
stats.probplot(data2, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(data2)
```

o valor de p está abaixo do valor de alfa(0.05), hipotese alternativa aceita
